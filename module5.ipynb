Import Required libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Setting the visual style
sns.set_theme(style="whitegrid")

# Loading the 'tips' dataset (standard for EDA practice)
df = sns.load_dataset('tips')
df.head()
```

```python
#create Histogram
plt.figure(figsize=(8, 5))
sns.histplot(df['total_bill'], kde=True, color='skyblue')
plt.title('Distribution of Total Bills')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.show()
```

###Interpretation:
The data is right-skewed, meaning most customers spend between $10 and $25. There are fewer instances of very high bills (over $40), suggesting that high-spending "whales" are the exception rather than the rule at this establishment.

```python
#create Boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(x='day', y='total_bill', data=df, palette='Set2')
plt.title('Bill Distribution by Day of the Week')
plt.show()
```

### Interpretation:
 We can see that the median bill is higher on weekends (Saturday and Sunday) compared to weekdays. Additionally, Saturday shows several outliers (dots above the whiskers), indicating a few unusually high-spending tables that day.

```python
#create Correlation Heatmap
plt.figure(figsize=(8, 5))
# Only numeric columns for correlation
corr = df[['total_bill', 'tip', 'size']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

### Interpretation:
 There is a strong positive correlation (approx. 0.68) between total_bill and tip. This confirms the logical intuition: as the bill amount increases, the tip amount tends to increase as well.

```python
#create Scatter Plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x='total_bill', y='tip', hue='time', data=df)
plt.title('Relationship between Total Bill and Tip')
plt.show()
```

### Interpretation:
 The upward trend confirms the positive correlation. By adding hue='time', we can see that dinner bills (often higher) and lunch bills follow a similar tipping slope, though dinner provides more opportunities for higher absolute tips.

```python
#create Bar Chart
plt.figure(figsize=(8, 5))
sns.barplot(x='day', y='tip', data=df, estimator='mean', palette='viridis')
plt.title('Average Tip Amount per Day')
plt.show()
```

### Interpretation:
 While Sunday has a higher median bill (seen in the boxplot), the average tip is also highest on Sundays. This suggests that Sunday customers might be more generous or are dining in larger groups that tip better.
